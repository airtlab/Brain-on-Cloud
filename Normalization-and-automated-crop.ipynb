```
import os
import nibabel as nib
import numpy as np
```

```
def normalize(img):
  mean = img.mean() 
  std = img.std() 
  return (img - mean) / std
```

### OASIS-3

```
files = []
path = '.../' # AD
for r, d, f in os.walk(path):
  for file in f:
    if file.endswith('.nii.gz'):
      files.append(os.path.join(r, file))
```

```
i = 0
for file in range (len(files)):
  scan = nib.load(files[i]) 
  nii_data = scan.get_fdata() 
  nii_data_norm = normalize(nii_data)
  scan = nii_data_norm.astype(int) 
  nonzeros = np.where(scan != 0)
  scan_new = scan[min(nonzeros[0]) : max(nonzeros[0]) + 1, min(nonzeros[1]) : max(nonzeros[1]) + 1] 
  cropped_scan = []
  for img in scan_new:
    if not np.any(np.isnan(img)) and np.all(np.isfinite(img)): 
      cropped_scan.append(img)
  cropped_scan = np.array(cropped_scan) 
  #print("Shape before automated crop\t"+str(nii_data_norm.shape)+"") # x, y, z
  #print("Shape after automated crop\t"+str(cropped_scan.shape)+"") # x, y, z
  i+=1
  np.save(".../scan-"+str(i), cropped_scan) # choose your own path to store normalized and cropped OASIS-3 AD data
```
